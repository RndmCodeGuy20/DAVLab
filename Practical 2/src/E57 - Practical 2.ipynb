<h3 style="font-family: 'Inria Serif'; font-size:24px">2.1 Read the three csv files which contains the score of same students in term1 for each Subject</h3>

```python
maths_df = pd.read_csv('../data/MathsT1.csv')
physics_df = pd.read_csv('../data/PhysicsT1.csv')
ds_df = pd.read_csv('../data/DataStructuresT1.csv')
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.2 Remove the name and ethnicity column (to ensure confidentiality)</h3>

```python
maths_df.drop(['Name', 'Ethnicity'], axis=1, inplace=True)
physics_df.drop(['Name', 'Ethnicity'], axis=1, inplace=True)
ds_df.drop(['Name', 'Ethnicity'], axis=1, inplace=True)
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.3 Fill missing score data with zero</h3>

```python
merged_df = pd.merge(maths_df, physics_df, on='ID')
```

```python
merged_df.head()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.5 Change Sex (M/F) Column to 1/2 for further analysis</h3>

```python
maths_df['Sex'] = maths_df['Sex'].replace({'M': 1, 'F': 2})

maths_df.head()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.6 Store the data in new file – ScoreFinal.csv</h3>

```python
merged_df.to_csv('../data/ScoreFinal.csv', index=False)
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.7 Convert ethnicity to numerical value</h3>

```python
maths_df = pd.read_csv('../data/MathsT1.csv')

maths_df['Ethnicity'] = maths_df['Ethnicity'].replace({'White': 1, 'Black': 2, 'Hispanic': 3, 'Asian': 4})
```

```python
maths_df.head()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">2.8 Fill the missing score for a student to the average of the class</h3>

```python
maths_df['Score'].fillna(maths_df['Score'].mean(), inplace=True)
```

<h2 style="font-family: 'Inria Serif'; font-size:34px">3. Source Code - Part B</h2>

```python
sales_df = pd.read_csv('../data/sales_data.csv', encoding='latin1')
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">3.1 Use barchart to plot the sales of different product lines each year and present the analysis.</h3>

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='YEAR_ID', y='SALES', hue='PRODUCTLINE', data=sales_df, palette=colors)

plt.title('Sales of different product lines each year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.show()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">3.2 Plot boxplot for total sales values for each year and present the analysis.</h3>

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='YEAR_ID', y='SALES', data=sales_df, palette=colors)

plt.title('Total sales values for each year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.show()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">3.3 Plot scatter plot for total sales value for each year and present the analysis</h3>

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x='SALES', y='YEAR_ID', data=sales_df, hue='PRODUCTLINE', palette=colors)

plt.title('Total sales value for each year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.show()
```

<h3 style="font-family: 'Inria Serif'; font-size:24px">3.4 Generate histogram for total sales monthwise and yearwise</h3>

```python
plt.figure(figsize=(12, 8))

sns.histplot(x='MONTH_ID', data=sales_df, hue='YEAR_ID', palette=colors)

plt.title('Total sales monthwise and yearwise')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```
